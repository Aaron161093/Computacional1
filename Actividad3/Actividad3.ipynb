```python
x = df0.PRES
y = df0.HGHT

plt.figure()
plt.plot(x, y)
 
plt.xlabel('PRESION (Pa)')
plt.ylabel('ALTURA (m)')
 
plt.title('Presion como funcion de la altura')

plt.grid(True)
 
plt.show()
```
